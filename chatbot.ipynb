```python
# Importer les bibliothèques nécessaires
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from pycaret.classification import *
```

```python
# Charger les données depuis le fichier JSON
with open('intents.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    
data    
```

```python
# Convertir le JSON en DataFrame
df = pd.DataFrame(data['intents'])
df
```

```python
# Extraire les patterns et les tags pour PyCaret
patterns = []
tags = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        patterns.append(pattern)
        tags.append(intent['tag'])
```

```python
# Créer un DataFrame pour les données d'entraînement
data_df = pd.DataFrame({'text': patterns, 'label': tags})
```

```python
# Diviser les données en ensembles d'entraînement et de test
train_df, test_df = train_test_split(data_df, test_size=0.2, random_state=42)
```

```python
# Configurer PyCaret pour la classification
# Cette étape transforme automatiquement les labels en valeurs numériques
clf_setup = setup(
    data=train_df,
    target='label',          # La colonne cible
    text_features=['text'],   # Colonne contenant les données textuelles
    session_id=42,
    verbose=False
)
```

```python
# Comparer les modèles pour choisir le meilleur
best_model = compare_models()
```
